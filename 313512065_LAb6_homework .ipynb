```python
# --- Binary Image Classification Colab Notebook (CNN with Redesign) ---

# ✅ 1. Import Libraries
import pandas as pd
import IPython.display as display
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import os, random
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

print("TensorFlow version:", tf.__version__)

# ✅ 2. Download and Unzip Dataset
!pip install -U gdown
import gdown, zipfile

file_id = "1eyrNGFlM83pf-TETo30Cvjm7kYuCFAqu"
gdown.download(f"https://drive.google.com/uc?id={file_id}", output="dataset.zip", quiet=False)

with zipfile.ZipFile("dataset.zip", 'r') as zip_ref:
    zip_ref.extractall("dataset")

# ✅ 3. Dataset Preparation
IMAGE_HEIGHT = 128
IMAGE_WIDTH = 128
BATCH_SIZE = 64

def get_pathframe(path):
    filenames = os.listdir(path)
    categories = []
    paths = []
    for filename in filenames:
        paths.append(path + filename)
        category = filename.split('.')[0]
        if category == 'dog':
            categories.append(1)
        else:
            categories.append(0)
    df = pd.DataFrame({
        'filename': filenames,
        'category': categories,
        'paths': paths
    })
    return df

df = get_pathframe("dataset/dataset/dataset/")
df.tail()

# ✅ 4. Convert to Tensor
def load_and_preprocess_image(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [IMAGE_WIDTH, IMAGE_HEIGHT])
    image /= 255.0
    return image

def convert_to_tensor(df):
    path_ds = tf.data.Dataset.from_tensor_slices(df['paths'])
    image_ds = path_ds.map(load_and_preprocess_image)
    label_ds = tf.data.Dataset.from_tensor_slices(tf.cast(df['category'], tf.int64))
    return image_ds, label_ds

X, Y = convert_to_tensor(df)

# ✅ 5. Train/Test Split and Plot Samples
dataset = tf.data.Dataset.zip((X, Y)).shuffle(buffer_size=2000)
dataset_train = dataset.take(22500).batch(BATCH_SIZE, drop_remainder=True)
dataset_test = dataset.skip(22500).batch(BATCH_SIZE, drop_remainder=True)

def plotimages(imagesls):
    fig, axes = plt.subplots(1, 5, figsize=(20, 20))
    axes = axes.flatten()
    for image, ax in zip(imagesls, axes):
        ax.imshow(image)
        ax.axis('off')
    plt.show()

imagesls = []
for image in X.take(5):
    imagesls.append(image)
plotimages(imagesls)

# ✅ 6. Redesigned CNN Model
def Augmented_CNNmodel():
    model = tf.keras.models.Sequential([
        layers.InputLayer(input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, 3)),

        # Data Augmentation
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),

        layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),

        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),

        layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(0.3),

        layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(0.4),

        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])

    opt = tf.keras.optimizers.Adam(learning_rate=0.0005)
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = Augmented_CNNmodel()
model.summary()

# ✅ 7. Train the Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("best_model.keras", save_best_only=True)
]

hist = model.fit(dataset_train, epochs=20, validation_data=dataset_test, callbacks=callbacks)

```

```python
# ✅ 7. Train the Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("best_model.keras", save_best_only=True)
]

hist = model.fit(dataset_train, epochs=20, validation_data=dataset_test, callbacks=callbacks)
```

```python
# ✅ 8. Plot Training History
def plot_model_history(model_history, acc='accuracy', val_acc='val_accuracy'):
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))

    axs[0].plot(model_history.history[acc])
    axs[0].plot(model_history.history[val_acc])
    axs[0].set_title('Model Accuracy')
    axs[0].set_ylabel('Accuracy')
    axs[0].set_xlabel('Epoch')
    axs[0].legend(['Train', 'Validation'], loc='best')

    axs[1].plot(model_history.history['loss'])
    axs[1].plot(model_history.history['val_loss'])
    axs[1].set_title('Model Loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].legend(['Train', 'Validation'], loc='best')

    plt.tight_layout()
    plt.show()

plot_model_history(hist)

# ✅ 9. Evaluate Model
loss, accuracy = model.evaluate(dataset_test)
print("Test Accuracy: {:.4f}, Loss: {:.4f}".format(accuracy, loss))

# ✅ 10. Confusion Matrix & Classification Report
y_pred = model.predict(dataset_test)
y_p = np.where(y_pred > 0.5, 1, 0).flatten()

test_data = dataset_test.unbatch()
y_g = [label.numpy() for _, label in test_data]
y_g = np.array(y_g).flatten()

confusion_mtx = confusion_matrix(y_g, y_p)

f, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(confusion_mtx, annot=True, linewidths=0.01, cmap="Blues", linecolor="gray", fmt='.1f', ax=ax)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

report = classification_report(y_g, y_p, target_names=['Cat (0)', 'Dog (1)'])
print(report)
```

**Model Design Rationale**
In this project, the model was designed to progressively extract hierarchical features by stacking four convolutional blocks with increasing filter sizes (32 → 64 → 128 → 256). Each convolutional layer is followed by Batch Normalization to stabilize and accelerate training, while Dropout with increasing rates (0.3, 0.4, 0.5) was applied to reduce overfitting. Data augmentation techniques such as RandomFlip, RandomRotation, and RandomZoom were incorporated to enhance data diversity and improve the model's robustness. The Adam optimizer with a custom learning rate of 0.0005 was used for efficient optimization. Overall, the design balances model capacity and regularization, leading to stable convergence and improved validation accuracy.

**Performance Improvement Explanation**
Compared to the baseline model, the redesigned architecture achieved significantly better validation accuracy. The main contributors to this improvement include the use of deeper convolutional layers for capturing richer features, Batch Normalization for stabilizing training, and progressively larger Dropout rates to combat overfitting. Additionally, data augmentation techniques introduced more variations in the training set, helping the model generalize better to unseen data. EarlyStopping and ModelCheckpoint callbacks further ensured that the best performing model was preserved without overtraining. Together, these strategies led to faster convergence, reduced validation loss, and a noticeable accuracy gain compared to the original model.
